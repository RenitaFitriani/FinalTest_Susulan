```python
# Nama : Renita Fitriani
# Soal : Essay 1 - ETL

import requests
import pandas as pd

# 1. API Request
api_url = "https://berita-indo-api-next.vercel.app/api/cnn-news/teknologi"
response = requests.get(api_url)

# Periksa apakah permintaan berhasil
if response.status_code == 200:
    # Dapatkan data JSON dari respons
    api_data = response.json()

    # Cetak pesan dan total berita
    print(api_data["message"])
    print(f"Total Berita: {api_data['total']}")

# 2. Transformasi menjadi DataFrame
    data_list = api_data["data"]

    # Buat DataFrame dari list data
    df = pd.DataFrame(data_list)

    # Tampilkan DataFrame
    print("\nDataFrame:")
    print(df)

# 3. Transformasi data kolom 'isoDate' menjadi tipe data datetime
    df['isoDate'] = pd.to_datetime(df['isoDate'])

    # Cetak DataFrame setelah konversi 'isoDate' ke tipe data datetime
    print("\nDataFrame setelah konversi isoDate:")
    print(df)

# 4. Aggregasi data berdasarkan kolom 'title' dan urutkan berdasarkan jumlah berita terbanyak
    aggregation_result = df.groupby('title').size().reset_index(name='jumlah_berita')
    sorted_result = aggregation_result.sort_values(by='jumlah_berita', ascending=False)

    # Tampilkan hasil
    print("\nHasil Agregasi:")
    print(sorted_result)

else:
    print(f"Error {response.status_code}: Gagal mendapatkan data dari API")
```
